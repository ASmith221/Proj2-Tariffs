```python
# Dependencies
from sqlalchemy import create_engine
import pandas
```

```python
# make connection to the SQLite database
engine = create_engine('sqlite:///Data/data.sqlite')
conn = engine.connect()
```

```python
# read import data from Excel file
df = pandas.read_excel('Data/Imports.xlsx')
df.head()
```

```python
# using pandas to insert the Imports table in to SQLite database
df.to_sql('Imports', conn,if_exists='append', index=False)
```

```python
# read export data from Excel file
df = pandas.read_excel('Data/Exports.xlsx')
df.head()
```

```python
# using pandas to insert the Imports table in to SQLite database
df.to_sql('Exports', conn,if_exists='append', index=False)
```
