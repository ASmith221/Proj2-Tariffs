```python
# Dependencies
from sqlalchemy import create_engine
import pandas
```

```python
# make connection to the SQLite database
engine = create_engine('sqlite:///Data/data.sqlite')
conn = engine.connect()
```

```python
# read import data from Excel file
df = pandas.read_excel('Data/Imports.xlsx')
df.head()
```

```python
# using pandas to insert the Imports table in to SQLite database
df.to_sql('Imports', conn,if_exists='replace', index=False)
```

```python
import os

import pandas as pd
import numpy as np

import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine

from flask import Flask, jsonify, render_template
from flask_sqlalchemy import SQLAlchemy

app = Flask(__name__)
```

```python
#################################################
# Database Setup
#################################################

app.config["SQLALCHEMY_DATABASE_URI"] = "sqlite:///Data/data.sqlite"
db = SQLAlchemy(app)

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(db.engine, reflect=True)

# # Save references to each table
# # Samples_Metadata = Base.classes.Exports
# # Samples = Base.classes.Imports
# exports = Base.classes.Exports
# imports = Base.classes.Imports
```
